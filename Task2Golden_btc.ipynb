```python
df['year'] =pd.DatetimeIndex(df['Date']).year
df['month'] =pd.DatetimeIndex(df['Date']).month
```

```python
df.head()
```

```python
df['month'].unique()
```

```python
df.info()
```

```python
btc_df=pd.DataFrame(df.groupby('year')['BTC'].mean())
btc_df
btc_df.index
plt.plot(btc_df.index,btc_df['BTC'])
```

```python
amzn_df=pd.DataFrame(df.groupby('year')['AMZN'].mean())
amzn_df
amzn_df.index
plt.plot(amzn_df.index,amzn_df['AMZN'])
```

```python
dpz_df=pd.DataFrame(df.groupby('year')['DPZ'].mean())
dpz_df
dpz_df.index
plt.plot(dpz_df.index,dpz_df['DPZ'])
```

```python
nflx_df=pd.DataFrame(df.groupby('year')['NFLX'].mean())
nflx_df
nflx_df.index
plt.plot(nflx_df.index,nflx_df['NFLX'])
```

# Year wise

```python
plt.plot(amzn_df.index,amzn_df['AMZN'],color='red')
plt.plot(btc_df.index,btc_df['BTC'],color='green')
plt.plot(nflx_df.index,nflx_df['NFLX'],color='yellow')
plt.plot(dpz_df.index,dpz_df['DPZ'],color='blue')
plt.legend(['AMZN', 'BTC', 'NFLX', 'DPZ'])
plt.show()
```

# Month

```python
btc_df2=pd.DataFrame(df.groupby('month')['BTC'].mean())
amzn_df2=pd.DataFrame(df.groupby('month')['AMZN'].mean())
dpz_df2=pd.DataFrame(df.groupby('month')['DPZ'].mean())
nflx_df2=pd.DataFrame(df.groupby('month')['NFLX'].mean())
```

```python
plt.plot(amzn_df2.index,amzn_df2['AMZN'],color='red')
plt.plot(btc_df2.index,btc_df2['BTC'],color='green')
plt.plot(nflx_df2.index,nflx_df2['NFLX'],color='yellow')
plt.plot(dpz_df2.index,dpz_df2['DPZ'],color='blue')
plt.legend(['AMZN', 'BTC', 'NFLX', 'DPZ'])
plt.show()
```
